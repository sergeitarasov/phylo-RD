```python
from grayscott import GrayScott
import json
```

```python
with open('output/sim_BM.json') as user_file:
  parsed_json = json.load(user_file)

dim = 50 # number of dimensions
N = 50
Nt = 32000 
parsed_json.update({'dim':[N,N]})

U = []
V = []
for tip, values in parsed_json['tip_traits'].items():
  Du, Dv, F, K = values
  rdSolver = GrayScott(N)
  L = rdSolver.laplacian()
  rdSolver.initialise()
  rdSolver.integrate(Nt, Du, Dv, F, K, L)
  a, b = rdSolver.output()
  U.append(a.tolist())
  V.append(b.tolist())

parsed_json.update({'morphogen_U': U, 'morphogen_V':V})
```
